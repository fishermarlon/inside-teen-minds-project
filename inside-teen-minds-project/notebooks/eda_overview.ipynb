```python
df = pd.read_csv("../data/cleaned/cleaned_mental_health_main.csv")
```

```python
# Assign risk points based on conditions
df["risk_score"] = 0

# +2 points for high stress
df.loc[df["stress_level"] == "High", "risk_score"] += 2

# +1 point for low mood
df.loc[df["mood"] <= 3, "risk_score"] += 1

# +1 point for sleep less than 6 hours
df.loc[df["sleep_hours"] < 6, "risk_score"] += 1

# +1 point for low support (if scale is 1–5)
df.loc[df["support_feeling"] <= 2, "risk_score"] += 1
```

```python
# Create a new column called 'risk_group'
def classify_risk(score):
    if score >= 4:
        return "high_risk"
    elif score >= 2:
        return "moderate"
    else:
        return "low_risk"

df["risk_group"] = df["risk_score"].apply(classify_risk)
```

```python
df["risk_group"].value_counts()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="risk_group", order=["low_risk", "moderate", "high_risk"])
plt.title("Student Risk Segments")
plt.tight_layout()
plt.savefig("../outputs/charts/risk_group_counts.jpg", format="jpg", dpi=300)
plt.show()
```

## 🧩 Step 4 Summary: Risk Segmentation

Using a custom risk scoring model based on stress level, mood, sleep hours, and support feeling, each student was classified into a risk group:

- **High-Risk Students**: Scored 4 or more — showing multiple signs of emotional strain and reduced support
- **Moderate Risk**: Scored 2–3 — may benefit from proactive support or monitoring
- **Low Risk**: Scored 0–1 — currently stable across key wellness metrics

This segmentation allows schools or mental health programs to quickly identify which students may need immediate outreach and which groups can be used to study resilience factors.
